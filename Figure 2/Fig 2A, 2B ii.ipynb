```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Overlap of CCLE gene sets with Cancer Hallmark gene sets

```python
data=pd.read_excel("Overlap/CCLE_ref_Cancer_hallmark.xlsx",index_col=0)
data.head()
Gene_lists=["SCLC-A2","SCLC-A","SCLC-N","SCLC-P","SCLC-Y"]
Samples=["SCLC-A2","SCLC-A","SCLC-N","SCLC-P","SCLC-Y"]
print(data)
```

```python
data=data.to_numpy()
data
```

```python
sns.set(font_scale=1.5)
b=sns.heatmap(data,annot=True,
              annot_kws={'size': 18},linewidths=.5)
plt.xlabel("CCLE Gene Sets", fontsize=18)
plt.ylabel("Archetype Gene Sets", fontsize=18)
plt.title("Overlap Plot CCLE",fontsize=18)
plt.tick_params(axis='both', which='major', pad=4)
plt.yticks([0.1, 1.2, 2.2,3.2,4.3])
b.set_yticklabels(Samples, size = 10)
b.set_xticklabels(Samples, size = 10)
plt.savefig('CCLEvsCH_HeatMap.jpg',dpi=600)
```

## Overlap of GSE73160 gene sets with Cancer Hallmark gene sets

```python
data=pd.read_excel("Overlap/GSE73160_ref_Cancer_hallmark.xlsx",index_col=0)
data.head()
Gene_lists=["SCLC-A2","SCLC-A","SCLC-N","SCLC-P","SCLC-Y"]
Samples=["SCLC-A2","SCLC-A","SCLC-N","SCLC-P","SCLC-Y"]
print(data)
```
